```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:1234@127.0.0.1:3306/stock_db')

# 티커
query_ticker = """select * from kor_ticker;"""
kor_ticker = pd.read_sql(query_ticker, con=engine)
kor_ticker.to_pickle("data/kor_ticker.pkl")

# 주가
query_price = """select * from kor_price;"""
kor_price = pd.read_sql(query_price, con=engine)
kor_price.to_pickle("data/kor_price.pkl")

# 섹터
query_sector = """select * from kor_sector;"""
kor_sector = pd.read_sql(query_sector, con=engine)
kor_sector.to_pickle("data/kor_sector.pkl")

# 재무제표
query_fs = """select * from kor_fs;"""
kor_fs = pd.read_sql(query_fs, con=engine)
kor_fs.to_pickle("data/kor_fs.pkl")

# 밸류
query_value = """select * from kor_value;"""
kor_value = pd.read_sql(query_value, con=engine)
kor_value.to_pickle("data/kor_value.pkl")
```
